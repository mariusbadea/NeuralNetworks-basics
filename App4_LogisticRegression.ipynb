### Application 4: Logistic Regression

We generate random bi-dimensional training examples: x1, x2.<br>
We use the line W*X+b do define our outputs:
if (x1, x2) point is above the line then the output is 0; if below the line then output is 1
Our neuron must find the line that divides the training examples into 2 categories. This is called Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from nn_functions import nn_train

%matplotlib inline
```

```python
m = 200
W_train = np.array([3, -1], ndmin=2)
b_train = -0.5
X = np.random.rand(2, m)

print("X shape = " + str(X.shape))
```

```python
Y = (X[1] < (np.dot(W_train, X) + b_train)).astype(int)
```

```python
plt.scatter(X[0], X[1], c=Y[0])
plt.xlabel('X[0]', fontsize=14)
plt.ylabel('X[1]', fontsize=14)
```

```python
lr = 0.002
epochs = 20001
activation = 'sigmoid'
loss = 'logistic'
W = np.zeros((2, 1))
b = 0
```

```python
W, b = nn_train(X, Y, W, b, lr, epochs, activation, loss)
```

```python
plt.scatter(X[0], X[1], label='inputs', c=np.squeeze(Y))
Xplt = np.linspace(0,1)
plt.plot(Xplt, -(W[0]*Xplt + b)/W[1], label='neuron output function')
plt.xlabel('x', fontsize=14)
plt.ylabel('separation = ' + str(W_train) +' * X + ' + str(b_train) + '\n' + 
           'yhat = ' + str(np.round(W,2)) +' * X + ' + str(np.round(b,2)), fontsize=14, rotation=0, labelpad=120)
plt.legend()
```
