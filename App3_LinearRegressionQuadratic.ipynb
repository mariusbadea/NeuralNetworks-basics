### Application 3: Linear Regression with Quadratic Equation

We generate training examples with the formula: <br>
$ y = w_1 * x^2 + w_2 * x + b_t + random $<br>
Being provided x and y, our neuron must learn w1, w2, b. In other words, the neuron will find the curve w1*x^2 + w2*x + b that best approximate our training examples.

This time, instead of having 2 independent inputs, we have a quadratic equation. Everything else stays the same.<br>
As you have noticed, the function of neurons only compute a linear equation. So how is it possible for the neural networks to deal with quadratic equations? The only trick here is that we have to provide x<sup>2</sup> input to the network.

```python
import numpy as np
import matplotlib.pyplot as plt
from nn_functions import nn_train

%matplotlib inline
```

```python
m = 50
X = np.random.rand(1, m)*10
print("X.shape = " + str(X.shape))
```

```python
Xext = np.append(X*X, X, axis=0)
print("Xext.shape = " + str(Xext.shape))
```

```python
W_train = np.array([1, 2], ndmin=2)
b_train = 3

Y = np.dot(W_train, Xext) + b_train + np.random.randn(m)*10

print("W_train.shape = " + str(W_train.shape))
print("Y.shape = " + str(Y.shape))
```

```python
plt.scatter(X, Y)
plt.xlabel('x', fontsize=14)
plt.ylabel('Y = ' + str(W_train[0][0]) +' * X^2 + ' + str(W_train[0][1]) +' * X + ' + str(b_train) + ' + random', fontsize=14)
```

```python
W = np.zeros((2, 1))
b = 0
epochs = 5001
lr = 0.0003
activation = 'identity'
loss = 'mse'
```

```python
W, b = nn_train(Xext, Y, W, b, lr, epochs, activation, loss)
```

```python
plt.scatter(X, Y, label='inputs')
Xplt = np.arange(11)
plt.plot(Xplt, W[0]*Xplt*Xplt + W[1]*Xplt + b, label='neuron output function')
plt.xlabel('x', fontsize=14)
plt.ylabel('y = ' + str(W_train) +' * x + ' + str(b_train) + ' + random\n' + 
           'yhat = ' + str(np.round(W,2)) +' * x + ' + str(np.round(b,2)), fontsize=14, rotation=0, labelpad=120)
plt.legend()
```
