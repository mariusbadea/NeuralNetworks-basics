### Application 2: Linear Regression with Two Input Features

We generate training examples with the formula: <br>
$ y = w_1 * x_1 + w_2 * x_2 + b_t + random $<br>
Being provided x and y, our neuron must learn w1, w2, b. In other words, the neuron will find the plane w1*x1 + w2*x2 + b that best approximate our training examples.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from nn_functions import nn_train

%matplotlib inline
```

```python
m = 50
X = np.random.rand(2, m)*10
print("X.shape = " + str(X.shape))
```

```python
W_train = np.array([[1], [2]])
b_train = 3
Y = np.dot(W_train.T, X) + b_train + np.random.randn(m)

print("W_train.shape = " + str(W_train.shape))
print("Y.shape = " + str(Y.shape))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(X[:][0], X[:][1], Y, cmap='hsv')
```

```python
W = np.zeros((2, 1))
b = 0
epochs = 5001
lr = 0.002
activation = 'identity'
loss = 'mse'
```

```python
W, b = nn_train(X, Y, W, b, lr, epochs, activation, loss)
```

```python
#%matplotlib notebook
%matplotlib inline

fig = plt.figure(figsize=(6, 4))
ax = plt.axes(projection='3d')
ax.scatter3D(X[:][0], X[:][1], Y, cmap='hsv', label='inputs')

def f(x, y):
    return W[0]*x + W[1]*y + b

x = np.linspace(0, 10, 10)
y = np.linspace(0, 10, 10)

Xplt, Yplt = np.meshgrid(x, y)
Zplt = f(Xplt, Yplt)
ax.contour3D(Xplt, Yplt, Zplt, 50, cmap='binary', label='neuron output function')
ax.set_xlabel('x1', fontsize=14)
ax.set_ylabel('x2', fontsize=14)
ax.set_zlabel('Y = ' + str(W_train) +' * X + ' + str(b_train) + ' + random\n' + 
           'Yhat = ' + str(np.round(W,2)) +' * X + ' + str(np.round(b,2)), fontsize=14, labelpad=20)
ax.legend()
```
