###  Application 1: Linear Regression with One Input Feature

We generate training examples with the formula: <br>
$ y_i = w_t * x_i + b_t + random $<br>
Being provided x and y, our neuron must learn w, b. In other words, the neuron will find the line w*x + b that best approximate our training examples. This is called Linear Regression.

```python
import numpy as np
import matplotlib.pyplot as plt
from nn_functions import nn_train

%matplotlib inline
```

```python
m = 50
X = np.random.rand(1, m)*10 # (0..10) interval
print('X.shape = ' + str(X.shape))
```

```python
W_train = np.array([2], ndmin=2)
b_train = 3
Y = np.dot(W_train, X) + b_train + np.random.randn(m)*10
print("W_train.shape = " + str(W_train.shape))
print("Y.shape = " + str(Y.shape))
```

```python
plt.scatter(X, Y)
plt.xlabel('x', fontsize=14)
plt.ylabel('y = ' + str(W_train[0][0]) +' * x + ' + str(b_train) + ' + random', fontsize=14)
```

```python
W = np.zeros((1,1))
b = 0
epochs = 5001
lr = 0.001
activation = 'identity'
loss = 'mse'
```

```python
W, b = nn_train(X, Y, W, b, lr, epochs, activation, loss)
```

```python
plt.scatter(X, Y, label='inputs')
Xplt = np.linspace(0,10)
plt.plot(Xplt, W[0]* Xplt + b, label='neuron output function')
#plt.plot(X, *X+b, label='neuron output function')
plt.xlabel('x', fontsize=14)
plt.ylabel('y = ' + str(W_train[0][0]) +' * x + ' + str(b_train) + ' + random\n' + 
           'yhat = ' + str(np.round(W[0][0],2)) +' * x + ' + str(np.round(b,2)), fontsize=14, rotation=0, labelpad=100)
plt.legend()
```
